```python
# Query Parameters
    # Example Url: books.com?auther=MJDeMarco&title=Millionaire&price=249
    # urls have a list of attributes separated by '&'
    # everything that follows the '?' is a query parameter
    # query params are used to send additional info to server while making requests
```

```python
import requests
url='https://api.coinbase.com/v2/exchange-rates?currency=BTC'
```

```python
response=requests.get(url)
```

```python
response.json()
```

```python
response.json()['data']['rates']['USD']
```

```python
# we shouldn't do query params in url
url='https://api.coinbase.com/v2/exchange-rates?currency=BTC'
# the url should be simply the base url & params should be factored externally
url='https://api.coinbase.com/v2/exchange-rates'
```

```python
params={'currency':'BTC'}
response=requests.get(url,params=params)
```

```python
response.json()['data']['rates']['USD']
```
