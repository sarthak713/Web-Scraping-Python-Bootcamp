```python
# Selector Types
```

```python
import requests
url='https://en.wikipedia.org/wiki/Rare-earth_element'
response=requests.get(url)

from selectolax.parser import HTMLParser
tree=HTMLParser(response.text)
```

```python
# simple selector - single component in a selector
len(tree.css('span'))
len(tree.css('.mw-editsection-bracket'))
len(tree.css('#firstHeading'))
```

```python
# compound selector - two or more components
tree.css('p.some_class')    # p tag with class = .someclass
tree.css('p[class]')        # p tag with class attribute
```

```python
# complex selector - simple + compound selector, combined with combinator
tree.css("a > img[src*='upload.wiki.org']")
```

```python
# selector lists - list of selectors separated by ',' to target more than one selector at once
tree.css("a > img[src*='upload.wiki.org'], p, a")
```
